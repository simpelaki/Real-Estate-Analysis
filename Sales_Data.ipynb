```python
combined_cities = pd.concat([Seattle_df, Tampa_df, San_Jose_df, Phoenix_df, Austin_df])
combined_cities
```

```python
combined_cities_list = []
for year in range(2015, 2023):
    possible_short_names = [f'{year}-{month:02d}'
                            for month in range(1, 13)]
    year_columns = [full_name
                    for full_name in combined_cities.columns
                    if full_name[:7] in possible_short_names]
    year_columns = ['RegionName', 'StateName'] + year_columns
    combined_cities_list.append(combined_cities[year_columns])
combined_cities_list[0]
```

```python
combined_cities2 = pd.concat([combined_cities_list[0], combined_cities_list[1], combined_cities_list[2], combined_cities_list[3],  combined_cities_list[4]], axis=1)
combined_cities2
```

```python
combined_cities_post = pd.concat([combined_cities_list[5], combined_cities_list[6], combined_cities_list[7]], axis=1)
combined_cities_post
```
