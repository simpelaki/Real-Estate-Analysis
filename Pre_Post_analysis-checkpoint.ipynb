```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import datetime as dt

#Study data file read from
sales_data=pd.read_csv("C:/Users/A/OneDrive/Desktop/DataClass/project_1/project1_saledata.csv")
sales_data.head()
```

```python
# defining each cities individually

Seattle_df = sales_data[sales_data["RegionName"] == "Seattle, WA"]
Seattle_df
```

```python
Tampa_df = sales_data[sales_data["RegionName"] == "Tampa, FL"]
Tampa_df
```

```python
San_Jose_df = sales_data[sales_data["RegionName"] == "San Jose, CA"]
San_Jose_df
```

```python
Austin_df = sales_data[sales_data["RegionName"] == "Austin, TX"]
Austin_df
```

```python
Phoenix_df = sales_data[sales_data["RegionName"] == "Phoenix, AZ"]
Phoenix_df
```

```python
#Combining all cities

combined_cities = pd.concat([Austin_df, Seattle_df, San_Jose_df,
                            Phoenix_df, Tampa_df])
combined_cities
```

```python
combined_cities_list = []
for year in range(2015, 2023):
    possible_short_names = [f'{year}-{month:02d}'
                            for month in range(1, 13)]
    year_columns = [full_name
                    for full_name in combined_cities.columns
                    if full_name[:7] in possible_short_names]
    year_columns = ['RegionName'] + year_columns
    combined_cities_list.append(combined_cities[year_columns])
combined_cities_list[0]
```

```python
combined_cities_pre = pd.concat([combined_cities_list[0], combined_cities_list[1], combined_cities_list[2], combined_cities_list[3],  combined_cities_list[4]], axis=1)
combined_cities_pre
```

```python
combined_cities_post = pd.concat([combined_cities_list[5], combined_cities_list[6], combined_cities_list[7]], axis=1)
combined_cities_post
```

```python
pre_transposed = combined_cities_pre.transpose()
pre_transposed
```

```python
combined_cities_pre = pd.concat([combined_cities_list[0], combined_cities_list[1], combined_cities_list[2], combined_cities_list[3],  combined_cities_list[4]], axis=1)
combined_cities_pre
```

```python
cities_2015 = combined_cities_list[0].set_index("RegionName")
cities_2016 = combined_cities_list[1].set_index("RegionName")
cities_2017 = combined_cities_list[2].set_index("RegionName")
cities_2018 = combined_cities_list[3].set_index("RegionName")
cities_2019 = combined_cities_list[4].set_index("RegionName")
```

```python
combined_cities_pre = pd.concat([cities_2015, cities_2016, cities_2017,cities_2018, cities_2019], axis=1)
combined_cities_pre
```

```python
ax = pre_transposed.plot(kind="bar")
plt.bar("combined_cities_pre", height = 5, width = 1)
ax.set_xticks([2, 14, 26, 38, 50])
ax.set_xticklabels(["2015","2016", "2017", "2018","2019"])
#ax.get_legend().remove()
#ax.set_yticks([1, 5, 9, 13, 17])
#ax.set_yticklabels(['100000', '200000', '300000', '400000', '500000'])
plt.title("Median Price Per Year")
plt.show()
```
